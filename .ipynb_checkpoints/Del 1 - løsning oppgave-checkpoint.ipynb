Etterspørsel etter vare Q er gitt ved p = 500 – 2Q, der p er prisen på gode, og Q er kvantum etterspurt. Anta at en monopolist produserer denne varen og har kostnadsfunksjon c(Q) = Q2. Anta at monopolisten vil maksimere sin profitt.

i) Hvor mye produserer monopolbedriften og til hvilken pris? Forklar intuisjonen bak tilpasningen. 



```python
import sympy as sp
import numpy as np
from matplotlib import pyplot as plt
```

```python
def demand(Q):
    return (500-2*Q)
```

```python
def marginalrevenue(Q):
    return (500-4*Q)
```

```python
def marginalcost(Q):
    return (2*Q)
```

```python
Q=sp.symbols('Q', real=True, positive=True)
equ=sp.Eq(marginalrevenue(Q),marginalcost(Q))
equ
```

### Optimal kvantum og pris for monopolbedriften er:

```python
Q_equ=sp.solve(equ,Q)[0]
Q_equ
round(Q_equ,2)
```

```python
p_eq=demand(Q_equ)
p_eq
round(p_eq,2)
```

```python
#drawing 800 points in the interval 0.0 to 300
Q = np.linspace(1,300,300)

#drawing marginalrevenue
plt.plot(Q,marginalrevenue(Q),label='MR')

#drawing marginalcost
plt.plot(Q,marginalcost(Q),label='MC')
#drawing demand
plt.plot(Q,demand(Q),color='green',label='Etterspørsel')

#adding legend:
plt.legend(loc='lower right')

#setting the axis lables
plt.ylabel('Pris')
plt.xlabel('Enheter')

#setting the range of y values (the price) to be between 0 and 500, so that we can more
#easily see the equilibrium price
plt.ylim((-2,500))
```

ii) Anta at monopolisten vil maksimere sin profitt. Hvor mye produserer monopolbedriften og til hvilken pris? Forklar intuisjonen bak tilpasningen. 



```python
def supply(Q):
    return 2*Q
```

```python
# Vi finner markedlikevet der etterspørsel er lik tilbud
Q=sp.symbols('Q', real=True, positive=True)
equ=sp.Eq(demand(Q),supply(Q))
equ
```

### Optimal pris og kvantum blir da:

```python
Q_equ=sp.solve(equ,Q)[0]
Q_equ
```

```python
P_eq=demand(Q_equ)
P_eq
```

```python
#drawing 300 points in the interval 0.0 to 300 
00
Q = np.linspace(1,300,300)

#drawing supply
plt.plot(Q,supply(Q),label='Tilbud')

#drawing demand
plt.plot(Q,demand(Q),color='green',label='Etterspørsel')

#adding legend:
plt.legend(loc='lower right')

#setting the axis lables
plt.ylabel('Pris')
plt.xlabel('Enheter')

#setting the range of y values (the price) to be between 0 and 500, so that we can more
#easily see the equilibrium price
plt.ylim((-2,500))
```
